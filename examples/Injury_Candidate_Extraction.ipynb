Now we'll use CoreNLP via ddlite's SentenceParser to parse each sentence. DocParser can handle this too; we didn't really need that call above. This can take a little while, so if the example has already been run, we'll reload it.

```python
send_file = '/Users/Jerry/Documents/CMPS290H/Project/data/temp/news_saved_send_v1.pkl'
try:
    with open(send_file, 'rb') as f:
        sents = cPickle.load(f)
except:
    %time sents = dp.parseDocSentences()
    with open(send_file, 'w+') as f:
        cPickle.dump(sents, f)
```

```python
injury_dict=[line.rstrip() for line in open('/Users/Jerry/Documents/CMPS290H/Project/data/dictionary/injury_dic.tsv')]
injury_dm = DictionaryMatch(label='Injury part', dictionary=injury_dict, ignore_case=False)
```

```python
Injury_Entity = Entities(sents, injury_dm)
```

```python
verb_tags = ['VB','VBD','VBN','VBP','VBZ','VBG']
noun_tags = ['NN','NNS','NNPS','NNP']
verb_list = []
noun_list = []
sentence_list = []
window_size = 4
sentence_example_file = open('/Users/Jerry/Documents/CMPS290H/Project/data/injury_sentence_list.txt','w')
for i in xrange(Injury_Entity.num_candidates()):
    e = Injury_Entity[i]
    size = len(e.lemmas)
    sentence_example_file.write(' '.join(e.words) + '\n')
    for index in xrange(size):
        if e.poses[index] in verb_tags:
            left = max(index-window_size,0)
            right = min(index+window_size,size)
            verb = ' '.join(map(lambda i: e.lemmas[i], xrange(left,right)))
            verb_list.append(verb)
        if e.poses[index] in noun_tags:
            noun_list.append(e.lemmas[index])


```
